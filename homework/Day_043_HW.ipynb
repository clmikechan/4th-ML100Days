2. 假設總共有 N 筆資料，每棵樹用取後放回的方式抽了總共 N 筆資料生成，請問這棵樹大約使用了多少 % 不重複的原資料生成?
hint: 0.632 bootstrap


Ans1: 因為每棵樹都只有選取少少的幾個特徵, 所以應該不會長成太深的樹(跟總特徵數比)

Ans2: 對單一棵樹來說 一筆樣本沒被取到的機率是 $(1 - \frac{1}{N})^{N}$
因為每筆樣本沒被取到的機率都一樣, 所以對單一棵樹來說, 沒被取到的樣本占整體的 $(1 - \frac{1}{N})^{N}$. 義即被使用的樣本占整體的 $1 - (1 - \frac{1}{N})^{N}$

```python
def choose_rate(N):
    not_choose_rate = (1 - 1 / N) ** N
    return 1 - not_choose_rate
```

```python
for i in range(1, 10001):
    print(f"{(i):>5} 筆樣本, 平均取樣率 {choose_rate((i)) * 100:7.3f}%")
```

可以看出在資料筆數夠多的時候($\ge 492$ 筆), 平均取樣率大約都在 $63.2\%$ 左右
